## 4. Indexing & Slicing

### 1D Arrays

```python
a[0]        # First element
a[-1]       # Last element
a[1:4]      # Elements at indices 1, 2, 3
a[:3]       # First three elements
a[2:]       # From index 2 to the end
a[::2]      # Every second element
a[::-1]     # Reverse the array
```

### 2D Arrays (`array[row, column]`)

```python
A[0, 1]       # Single element (row 0, column 1)

A[0]          # First row
A[:, 0]       # First column

A[1:3]        # Rows 1 and 2
A[:2, 1:]     # First two rows, columns 1 onward

A[:, :]       # Entire array
```

### Boolean Indexing

```python
a[a < 2]          # Elements less than 2
a[a % 2 == 0]     # Even elements

A[A > 10]         # Elements greater than 10
```

### Fancy Indexing

```python
a[[3, 0, 2]]          # Select arbitrary elements

A[[1, 0], [0, 1]]     # Elements (1,0) and (0,1)

A[[0, 2]]             # Rows 0 and 2
A[:, [0, 2]]          # Columns 0 and 2
```

<img src="img/indexing.png" alt="Indexing"/>

````python
# 1D Slicing
c[0:2]       # Select elements at index 0 and 1
c[::-1]      # Reverse the 1D array

# 2D Slicing (matrix[row, column])
d[0:2, 1]    # Select items from rows 0 and 1 in column 1
d[:, 0]      # Select every item in column 0 (the entire first column)
d[:1]        # Select all items in row 0 (equivalent to b[0:1, :])

# Advanced Indexing
c[c < 2]     # Boolean Masking: Select elements from a less than 2
d[[1, 0], [0, 1]];  # Fancy Indexing: Selects elements at (1,0) and (0,1)
````

### Useful Axis Rules

For a matrix with shape `(rows, columns)`:

```python
axis=0   # Operate down the rows (one result per column)

axis=1   # Operate across the columns (one result per row)
```

Example:

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6],
])

A.sum(axis=0)   # [5, 7, 9]
A.sum(axis=1)   # [6, 15]
```

# The 10-Minute Challenge: "The Calibration Crisis"
You are analyzing data from a particle detector grid, but the sensor calibration was misconfigured. 

````python
# Setup the Simulation
import numpy as np

# A 5x3 matrix of simulated sensor readings
readings = np.array([
    [12, 15, 18],
    [20, 24, 28],
    [14, 17, 20],
    [31, 35, 39],
    [10, 12, 14]
])

# Calibration offsets that need to be applied per column
offsets = np.array([2, 5, 10])
````

#### The Mission
Write three lines of code to fix the data and isolate anomalies without loops:

**Task A (Broadcasting):** Correct the data. Subtract offsets from readings so that column 0 loses 2, column 1 loses 5, and column 2 loses 10.

**Task B (Boolean Masking):** Create a filter to find all corrected values that are still strictly greater than 15.

**Task C (Extraction):** Extract those high-intensity values into a clean 1D array.
